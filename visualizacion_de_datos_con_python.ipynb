# Visualización de datos con Python: Matplotlib, Seaborn y Plotly

##Importaciones

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Datos

Se debe cargar el archivo de la carpeta [Data](https://github.com/Daimond92/python_data_science/tree/develop/Data)
- inmigrantes_canada.csv

```python
df = pd.read_csv('/content/inmigrantes_canada.csv')
```

### Conocer los datos

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

### Crear Dataframe

```python
df.set_index('Pais', inplace=True)
```

```python
df.head()
```

```python
anos = list(map(str, range(1980, 2014)))
anos
```

```python
colombia = df.loc['Colombia', anos]
colombia
```

```python
col_dict = {'Año': colombia.index.tolist(),
            'Inmigrantes': colombia.values.tolist()}
col_dict
```

```python
datos_col = pd.DataFrame(col_dict)
datos_col.tail()
```

## Crear gráfico

```python
plt.plot(datos_col['Año'], datos_col['Inmigrantes'])
```

### Edición del grafico de linea

```python
plt.plot(datos_col['Año'], datos_col['Inmigrantes'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010', '2014'])
plt.yticks([0, 500, 1000, 1500, 2000, 2500, 3000, 3500, 4000, 4500, 5000, 5500, 6000, 6500])
plt.show()
```

```python
plt.plot(datos_col['Año'], datos_col['Inmigrantes'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010', '2014'])
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.plot(datos_col['Año'], datos_col['Inmigrantes'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010', '2014'])
plt.show()
```

### Asignar titulo y rotulos

```python
plt.figure(figsize=(8, 4))
plt.plot(datos_col['Año'], datos_col['Inmigrantes'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010', '2014'])
plt.title('Inmigración de colombianos hacia Canada')
plt.xlabel('Año')
plt.ylabel('No. de Inmigrantes')
plt.show()
```

```python
df_comparacion = df_comparacion.T
```

```python
df_comparacion.tail()
```

```python
plt.plot(df_comparacion['Brasil'], label='Brasil')
plt.plot(df_comparacion['Argentina'], label='Argentina')
plt.title('Inmigración de brasileros y argentinos a Canadá')
plt.xlabel('Año')
plt.ylabel('Número de Inmigrantes')
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.legend()
plt.show()
```
