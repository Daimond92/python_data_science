## Gráfico de barras vertical

```python
colores = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown', 'slateblue',
           'gray', 'olive', 'navy', 'teal', 'tomato']
```

## Gráfico de barras horizontales

```python
fig, ax = plt.subplots(figsize=(10, 3))

ax.barh(sudamerica.index, sudamerica['Total'], color=colores)
ax.set_title('Inmigración de sudamericanos hacia Canadá\ndurante el periodo de 1980 hasta 2013',
             fontsize=18, loc='left')
ax.set_ylabel('')
ax.set_xlabel('No. de Inmigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

plt.show()
```

```python
sudamerica_sorted = sudamerica.sort_values('Total', ascending=True)
```

```python
fig, ax = plt.subplots(figsize=(10, 3))

ax.barh(sudamerica_sorted.index, sudamerica_sorted['Total'], color=colores)
ax.set_title('Inmigración de sudamericanos hacia Canadá\ndurante el periodo de 1980 hasta 2013',
             fontsize=18, loc='left')
ax.set_ylabel('')
ax.set_xlabel('No. de Inmigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

plt.show()
```

```python
colores = []
for pais in sudamerica_sorted.index:
    if pais == 'Colombia':
        colores.append('forestgreen')
    else:
        colores.append('silver')
```

```python
fig, ax = plt.subplots(figsize=(10, 3))

ax.barh(sudamerica_sorted.index, sudamerica_sorted['Total'], color=colores)
ax.set_title('Sudamérica: Colombia es el segundo país con más migrantes\n hacia Canadá durante el periodo 1980 - 2013',
             fontsize=16, loc='left')
ax.set_ylabel('')
ax.set_xlabel('No. de Inmigrantes', fontsize=13)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

plt.show()
```

### Añadir anotaciones

```python
fig, ax = plt.subplots(figsize=(10, 3))

ax.set_frame_on(False)
ax.get_xaxis().set_visible(False)
ax.tick_params(axis='y', which='both', size=0)

ax.barh(sudamerica_sorted.index, sudamerica_sorted['Total'], color=colores)
ax.set_title('Sudamérica: Colombia es el segundo país con más migrantes\n hacia Canadá durante el periodo 1980 - 2013',
             fontsize=16, loc='left')
ax.set_ylabel('')
#ax.set_xlabel('No. de Inmigrantes', fontsize=13)
#ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=10)

for i,j in enumerate(sudamerica_sorted['Total']):
    ax.text(j + 20, i, str(j), color='black', fontsize=9, ha='left', va='center')

plt.show()
```

## Segunda parte

Trabajas como Analista de Datos en una empresa minorista y te asignaron la tarea de crear una figura con subgráficos que muestre la variación en el número de ventas en cuatro tiendas diferentes a lo largo de un año. La gerencia de la empresa necesita visualizar claramente las tendencias de ventas en cada tienda, para que puedan tomar decisiones estratégicas sobre inventarios y acciones de marketing. Para ello, debes crear cuatro subgráficos dispuestos en dos filas y dos columnas, donde cada subgráfico represente una tienda diferente. En este desafío, cada subgráfico debe presentar un gráfico de líneas que muestre la variación del número de ventas a lo largo de los meses del año.

¡Ahora es el momento de mostrar tus habilidades en análisis de datos y visualización! Para crear el DataFrame con el número de ventas de las tiendas y crear la figura, utiliza la siguiente información:

```python
tiendas = ['A', 'B', 'C', 'D']
ventas_2022 = {'Ene': [100, 80, 150, 50],
    'Feb': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'May': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Sep': [240, 160, 290, 130],
    'Oct': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dec': [300, 350, 400, 250]}
```

**Consejo:** Para facilitar la creación de los subgráficos, puedes definir la columna "tiendas" como índice del DataFrame y utilizar la propiedad loc de la biblioteca Pandas para trazar cada una de las tiendas.

¡No olvides agregar un título general a la figura, títulos a los subgráficos y etiquetas a los ejes! Además, asegúrate del tamaño de la figura y del espaciado entre los subgráficos.

```python
df_ventas = pd.DataFrame(ventas_2022, index=tiendas)
```

```python
df_ventas
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 5))
fig.subplots_adjust(hspace=0.4, wspace=0.3)
fig.suptitle('Variación de ventas de las principales tiendas durante el año 2025')

axs[0, 0].plot(df_ventas.loc['A'])
axs[0, 0].set_title('Ventas de la tienda A')

axs[0, 1].plot(df_ventas.loc['B'])
axs[0, 1].set_title('Ventas de la tienda B')

axs[1, 0].plot(df_ventas.loc['C'])
axs[1, 0].set_title('Ventas de la tienda C')

axs[1, 1].plot(df_ventas.loc['D'])
axs[1, 1].set_title('Ventas de la tienda D')

ymin2 = 0
ymax2 = 450

for ax in axs.ravel():
    ax.set_ylim(ymin2, ymax2)

for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))
    ax.set_xlabel('Mes')
    ax.set_ylabel('No. de Ventas')
    ax.grid()

plt.show()
```

## Tercera parte

Algunas ideas de personalización para el gráfico de tiendas que puedes explorar son:

- Cambiar la posición de los títulos de los subgráficos a la izquierda.
- Aumentar el tamaño de la fuente del título general de la figura para resaltarlo.
- Aumentar el tamaño de los títulos y etiquetas de los ejes de los subgráficos.

**consejo:** puedes reducir el tamaño del código utilizando el comando for i, ax in enumerate(axs.flat):, que permite un bucle que itera sobre todos los subgráficos de la figura. Dentro de este bucle, puedes pasar las funciones plot, set_title, set_xlabel, set_ylabel, y así sucesivamente...

```python
colores = ['darkviolet', 'green', 'darkblue', 'coral']
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 5))
fig.subplots_adjust(hspace=0.6, wspace=0.4)
fig.suptitle('Variación de ventas de las principales tiendas durante el año 2025', fontsize=18)

ymin2 = 0
ymax2 = 450


for ax in axs.ravel():
    ax.set_ylim(ymin2, ymax2)

for i, ax in enumerate(axs.flat):
    ax.plot(df_ventas.loc[df_ventas.index[i]], color=colores[i], lw=3)
    ax.set_title(f'Ventas en la tienda {df_ventas.index[i]}', loc='left', fontsize=14)
    ax.set_xlabel('Mes', fontsize=12)
    ax.set_ylabel('Número de ventas', fontsize=12)
    ax.tick_params(labelsize=9)
    ax.grid(color='lightgrey')

plt.show()
```
