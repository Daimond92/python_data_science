# Visualización de datos con Python: Matplotlib, Seaborn y Plotly

##Importaciones

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Datos

Se debe cargar el archivo de la carpeta [Data](https://github.com/Daimond92/python_data_science/tree/develop/Data)
- inmigrantes_canada.csv

```python
df = pd.read_csv('/content/inmigrantes_canada.csv')
```

### Conocer los datos

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

### Crear Dataframe

```python
df.set_index('Pais', inplace=True)
```

```python
df.head()
```

```python
anos = list(map(str, range(1980, 2014)))
anos
```

```python
colombia = df.loc['Colombia', anos]
colombia
```

```python
col_dict = {'Año': colombia.index.tolist(),
            'Inmigrantes': colombia.values.tolist()}
col_dict
```

```python
datos_col = pd.DataFrame(col_dict)
datos_col.tail()
```

## Crear gráfico

```python
plt.plot(datos_col['Año'], datos_col['Inmigrantes'])
```

### Edición del grafico de linea

```python
plt.plot(datos_col['Año'], datos_col['Inmigrantes'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010', '2014'])
plt.yticks([0, 500, 1000, 1500, 2000, 2500, 3000, 3500, 4000, 4500, 5000, 5500, 6000, 6500])
plt.show()
```

```python
plt.plot(datos_col['Año'], datos_col['Inmigrantes'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010', '2014'])
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.plot(datos_col['Año'], datos_col['Inmigrantes'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010', '2014'])
plt.show()
```

### Asignar titulo y rotulos

```python
plt.figure(figsize=(8, 4))
plt.plot(datos_col['Año'], datos_col['Inmigrantes'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010', '2014'])
plt.title('Inmigración de colombianos hacia Canada')
plt.xlabel('Año')
plt.ylabel('No. de Inmigrantes')
plt.show()
```

## Creando almacenando en una variable

### Creando lienzo

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(datos_col['Año'], datos_col['Inmigrantes'])
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
plt.show()
```

### Asignar titulos y rotulos

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(datos_col['Año'], datos_col['Inmigrantes'])
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('Inmigración de colombianos hacia Canada\n1980 - 2013')
ax.set_xlabel('Año')
ax.set_ylabel('No. de Inmigrantes')
plt.show()
```

## Crear subplots análisis estadistico

### Crear gráficos en una dirección (1 fila y 2 columnas)

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 3))
axs[0].plot(datos_col['Año'], datos_col['Inmigrantes'])
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].set_title('Inmigración de colombianos hacia Canada\n1980 - 2013')
axs[0].set_xlabel('Año')
axs[0].set_ylabel('No. de Inmigrantes')
axs[0].grid()

axs[1].boxplot(datos_col['Inmigrantes'])
axs[1].set_title('Boxplot de la inmigración de colombianos \nhacia Canada 1980 - 2013')
axs[1].set_xlabel('Colombia')
axs[1].set_ylabel('No. de Inmigrantes')
axs[1].grid()

plt.show()
```

### Crear gráficos en 2 direcciones (2 filas y 2 columnas)

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 6))

axs[0, 0].plot(df.loc['Colombia', anos])
axs[0, 0].set_title('Colombia')

axs[0, 1].plot(df.loc['Brasil', anos])
axs[0, 1].set_title('Brasil')

axs[1, 0].plot(df.loc['Argentina', anos])
axs[1, 0].set_title('Argentina')

axs[1, 1].plot(df.loc['Perú', anos])
axs[1, 1].set_title('Perú')

plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)

axs[0, 0].plot(df.loc['Colombia', anos])
axs[0, 0].set_title('Colombia')

axs[0, 1].plot(df.loc['Brasil', anos])
axs[0, 1].set_title('Brasil')

axs[1, 0].plot(df.loc['Argentina', anos])
axs[1, 0].set_title('Argentina')

axs[1, 1].plot(df.loc['Perú', anos])
axs[1, 1].set_title('Perú')

plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Inmigración de los 4 países mayores sudamericanos \nhacia Canadá desde 1980 - 2013')

axs[0, 0].plot(df.loc['Colombia', anos])
axs[0, 0].set_title('Colombia')

axs[0, 1].plot(df.loc['Brasil', anos])
axs[0, 1].set_title('Brasil')

axs[1, 0].plot(df.loc['Argentina', anos])
axs[1, 0].set_title('Argentina')

axs[1, 1].plot(df.loc['Perú', anos])
axs[1, 1].set_title('Perú')

ymin = 0
ymax = 7000

for ax in axs.ravel():
    ax.set_ylim(ymin, ymax)

for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))
    ax.set_xlabel('Año')
    ax.set_ylabel('No. de Inmigrantes')
    ax.grid()

plt.show()
```

## Segunda parte

Trabajas como Analista de Datos en una empresa minorista y te asignaron la tarea de crear una figura con subgráficos que muestre la variación en el número de ventas en cuatro tiendas diferentes a lo largo de un año. La gerencia de la empresa necesita visualizar claramente las tendencias de ventas en cada tienda, para que puedan tomar decisiones estratégicas sobre inventarios y acciones de marketing. Para ello, debes crear cuatro subgráficos dispuestos en dos filas y dos columnas, donde cada subgráfico represente una tienda diferente. En este desafío, cada subgráfico debe presentar un gráfico de líneas que muestre la variación del número de ventas a lo largo de los meses del año.

¡Ahora es el momento de mostrar tus habilidades en análisis de datos y visualización! Para crear el DataFrame con el número de ventas de las tiendas y crear la figura, utiliza la siguiente información:

```python
tiendas = ['A', 'B', 'C', 'D']
ventas_2022 = {'Ene': [100, 80, 150, 50],
    'Feb': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'May': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Sep': [240, 160, 290, 130],
    'Oct': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dec': [300, 350, 400, 250]}
```

**Consejo:** Para facilitar la creación de los subgráficos, puedes definir la columna "tiendas" como índice del DataFrame y utilizar la propiedad loc de la biblioteca Pandas para trazar cada una de las tiendas.

¡No olvides agregar un título general a la figura, títulos a los subgráficos y etiquetas a los ejes! Además, asegúrate del tamaño de la figura y del espaciado entre los subgráficos.

```python
df_ventas = pd.DataFrame(ventas_2022, index=tiendas)
```

```python
df_ventas
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 5))
fig.subplots_adjust(hspace=0.4, wspace=0.3)
fig.suptitle('Variación de ventas de las principales tiendas durante el año 2025')

axs[0, 0].plot(df_ventas.loc['A'])
axs[0, 0].set_title('Ventas de la tienda A')

axs[0, 1].plot(df_ventas.loc['B'])
axs[0, 1].set_title('Ventas de la tienda B')

axs[1, 0].plot(df_ventas.loc['C'])
axs[1, 0].set_title('Ventas de la tienda C')

axs[1, 1].plot(df_ventas.loc['D'])
axs[1, 1].set_title('Ventas de la tienda D')

ymin2 = 0
ymax2 = 450

for ax in axs.ravel():
    ax.set_ylim(ymin2, ymax2)

for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))
    ax.set_xlabel('Mes')
    ax.set_ylabel('No. de Ventas')
    ax.grid()

plt.show()
```
