# Datos Númericos

## Retirar listas

```python
columnas = list(datos.columns)
columnas
```

```python
datos = datos.explode(columnas[3:])
```

```python
datos
```

```python
datos.reset_index(drop=True, inplace=True)
```

```python
datos.head()
```

```python
datos.info()
```

## Convertir a datos numericos

```python
datos['max_hospedes'] = datos['max_hospedes'].astype(np.int64)
```

```python
datos.info()
```

```python
columnas_numericas = ['cantidad_baños', 'cantidad_cuartos', 'cantidad_camas']
```

```python
datos[columnas_numericas] = datos[columnas_numericas].astype(np.int64)
```

```python
datos.info()
```

### Convertir a datos float

```python
datos['evaluacion_general'] = datos['evaluacion_general'].astype(np.float64)
```

```python
datos.info()
```

### Convertir número en string

```python
datos['precio'].astype(np.float64)
```

```python
datos['precio']
```

```python
datos['precio'] = datos['precio'].apply(lambda x: x.replace('$', '').replace(',', '').strip())
```

```python
datos['precio'] = datos['precio'].astype(np.float64)
```

```python
datos.info()
```

```python
datos[['cuota_deposito', 'cuota_limpieza']]
```

```python
datos[['cuota_deposito', 'cuota_limpieza']] = datos[['cuota_deposito', 'cuota_limpieza']].applymap(lambda x: x.replace('$', '').replace(',', '').strip())
```

```python
datos[['cuota_deposito', 'cuota_limpieza']] = datos[['cuota_deposito', 'cuota_limpieza']].astype(np.float64)
```

```python
datos.info()
```

## Desafio 1

### Parte 1

Link de archivo Ventas Online - [dados_vendas_clientes.json](https://cdn3.gnarususercontent.com.br/2928-transformacao-manipulacao-dados/dados_vendas_clientes.json)

```python
# URL del archivo JSON
url = "https://cdn3.gnarususercontent.com.br/2928-transformacao-manipulacao-dados/dados_vendas_clientes.json"

# Realizamos la solicitud GET para descargar el archivo
response = requests.get(url)

# Verificamos que la solicitud fue exitosa
if response.status_code == 200:
    # Guardamos el contenido en un archivo local, wb es write binary
    with open("dados_vendas_clientes.json", "wb") as file:
        file.write(response.content)
    print("Archivo descargado exitosamente.")
else:
    print(f"Error al descargar el archivo: {response.status_code}")
```
